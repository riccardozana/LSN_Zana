   L'obiettivo dell'esercizio √® stimare e graficare il valor medio e la relativa incertezza di una variabile pseudo-casuale r, $<r>$, che supponiamo essere distribuita uniformemente nell'intervallo $\lbrack 0; 1)$.
   La variabile pseudo-casuale √® stata generata attraverso il generatore di numeri casuali parallelo ("Parallel Random Number Generator") fornito.
   La stima dell'incertezza viene effettuata in funzione del numero $N$ di blocchi: proviene dal calcolo dei valori di 
    $$ \langle A^2 \rangle \simeq \frac{1}{N} \sum_{i=1}^N A^2_i \quad \quad \langle A\rangle^2 \simeq \left( \frac{1}{N} \sum_{i=1}^N A_i \right)^2 $$ come risultati di un esperimento, corrispondente al singolo blocco e infine calcolata come  $$\frac{\sigma}{\sqrt{N-1}} \simeq \sqrt{\frac{1}{N-1} \left[ \frac{1}{N} \sum_{i=1}^N A^2_i - \left( \frac{1}{N} \sum_{i=1}^N A_i \right)^2 \right]} $$
</p>
 
dove il valore di ognuno degli esperimenti √® stato ottenuto con $\frac{M}{N}$ punti, dove M √® il numero di punti totali e $N$ il numero di blocchi.
I valori scelti sono $M=10^4$ e $N=10^2$.
Si nota inoltre che dopo un solo esperimento non √® possibile avere una media dell'incertezza, perci√≤ il suo valore √® stato settato direttamente a 0.

### C. $\chi^2$

Infine vogliamo controllare se effettivamente l'ipotesi precedente, ovvero che i valori di r siano distribuiti uniformente, √® corretta. Per farlo calcoliamo il valore di $\chi^2$ dal quale √® possibile ricavare con quale livello di confidenza si rigetta l'ipotesi per la quale la distribuzione non √® uniforme. 
Per fare ci√≤ calcoliamo dividiamo l'intervallo $[0,1)$ in 100 sottointervalli e conteggiamo il numero di estrazioni in ogni sottointervallo.
Inoltre $\chi^2=‚àëùëñ(ùëÇùëñ‚àíùê∏ùëñ)2ùê∏ùëñ$ presenta al denominatore il valore di aspettazione del numero medio di conteggi in ogni intervallo siccome, per probabilit√† p piccole, dalla distibuzione binomiale si ricava che $\sigma^2 = np (1-p)$ e nel caso di p piccola si ottiene $\sigma^2 \simeq np$, che √® il valore di aspettazione. 
Il grafico propone il calcolo di 100 valori distinti di $\chi^2$, ognuno ottenuto con $10^4$ punti (n) e 100 sottocanali (M). 
$$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np

x,y=np.loadtxt("ex_1.01/chi2.out", usecols=(0, 1), unpack=True)

fig=plt.figure(figsize=(10, 7))
plt.errorbar(x,y)
plt.xlabel('Iterazioni')
plt.ylabel('$\chi^2$')
plt.hlines(100, 0, 100)
plt.title("Calcolo di $\chi^2$")
plt.grid(True)
plt.show()
```

Come si pu√≤ notare, il valore medio nel calcolo di $\chi^2$ √® pari a 100. Il risultato √® atteso siccome $(n_i - n/M)^2 \simeq n/M$ e perci√≤ $\chi^2 \simeq 100$. Se si fosse ottenuto un risultato maggiore, la il livello di confindenza con la quale si pu√≤ affermare che la distribuzione sia uniforme sarebbe stato molto basso. 

# 1.02
Il primo obiettivo dell'esercizio 2 √® ottenere due funzioni per generare numeri casuali con una distribuzione esponenziale e una lorentiana. 
    Per fare ci√≤ si utilizza il metodo dell'inversione della funzione cumulativa. Infatti, definendo una funzione cumulativa $F(x)=\int_{
   -\inf}^x p(x') dx' $, e una funzione y=y(x) scelta come $y=F_x(x)$, si ottiene a partire da $p_y(y)=p_x(x)\abs{(\frac{dF_x(x)}{dx})}$ si ottiene $p_y(y)=1$. Perci√≤, una volta che $F_x(x)$ √® nota e invertibile analiticamente, a partire da una variabile casuale distribuita uniformemente tra $[0,1)$ (ovvero y), √® possibile ottenere una variabile $x=F^{-1}(y)$ distribuita come $p_x(x)$. 
   
Da questo risultato possiamo quindi calcolare le funzioni cumulative della distribuzione esponenziale e lorentziana e ottere un seguito l'espressione di x in funzione di y.

##### Esponenziale: 
Sia $p_x(x)=\lambda exp(-\lambda x)$ con $0\le x < \inf $. Calcolando la funzione cumulativa si ottiene che $$F_x(x)=\int_0^x \lambda exp(-\lambda x) = 1-exp(-\lambda x)$$
dal qualla quale segue che la variabile estratta secondo una distibuzione esponenziale la si ottiene come $$x=-\frac{1}{\lambda}\log(1-y)$$.

##### Lorentziana:
Partendo da $p_x(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$
e seguendo il medesimo procedimento si ottiene che $$x=\Gamma \tan[\pi(\frac{1}{2}-y)]+\mu$$
