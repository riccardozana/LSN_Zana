E' possibile modellizzare l'andamento di titoli finanziari attraverso il Moto Browniano Geometrico. Infatti, definendo il valore di un titolo finanziario nel tempo come $S(t)$, la sua variazione sarà soggetta a due contributi. Il drift, $\mu$, misura il tasso di crescita medio del valore del titolo finanziario mentre la volatilità, $\sigma$, modellizza il contributo stocastico nell'evoluzione temporale del prezzo. Combinando i due contributi si ottiene la seguente equazione differenziale stocastica $dS(t)=\mu S(t)dt+\sigma S(t)dW(t)$ che definisce appunto una variante del Moto Browniano, il Moto Browniano Geometrico (GBM).
    
Siccome è di interesse la variazione relativa del prezzo, usando la formula di Ito si ottiene: $d \ln S(t)=(\mu-\frac{\sigma^2}{2})dt +\sigma dW(t)$, dalla quale per integrazione si ricava l'espressione esplicita $S(t)=S(0)exp((\mu-\frac{\sigma^2}{2})t +\sigma W(t))$. 
 
Il fine dell'esercizio è prezzare, attraverso il metodo montecarlo, un'opzione europea sia di tipo call che put e confrontare il risultato con i valori ricavati dall'espressione analitica esatta ricavata da Black e Sholes. 
    
Il parametri dati sono: 
- prezzo a $t=0$: $S(0)=100$
- tempo di consegna: $T=1$
- prezzo concordato: $K=100$
- tasso di interesse : $r=0.1$
- volatilità: $\sigma=0.25$

Stimiamo il prezzo delle opzioni in due modi: il primo attraverso l'espressione esatta $S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$, dove $W(t)$ ~ $\mathcal{N}(0, t)$ mentre nel secondo modo costruendo la traiettoria del processo stocastico come $S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$ con $Z_i$ ~ $\mathcal{N}(0,1)$, dividendo l'intervallo temporale $[0, T]$ in 100 sottointervalli (anche questo metodo risulta essere esatto, ovvero la probabilità congiunta al tempo $t_i$ corrisponde a quella ottenuta da $GBM(\mu, \sigma^2)$ ).
    
Una volta simulati i prezzi al tempo di consegna, $S(T)$, otteniamo la prezzatura di un'opzione call o put al tempo $t=0$, $C[S(0), 0]$ e $P[S(0), 0]$, partendo dal guadagno $G$ e riscalandolo di un fattore $\exp(-rT)$ (il quale modellizza il fatto che immobilizzando la somma di denaro in una banca avrei avuto un interesse al tempo finale pari a $\exp(rT)$ senza rischio). La funzione di guadagno $G$ per un'opzione call risulta essere $G:=max(0, S(T)-K)$, mentre $G:=max(0, K-S(T))$ per una put. 
    
In conclusione si è calcolato il valore medio di $C[S(0), 0]$ e $P[S(0), 0]$ attraverso $M=10^5$ valutazioni del prezzo $S(T)$ con ognuna delle due tecniche, divisi in seguito in $N=100$ blocchi per il calcolo dell'incertezza. 


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

###Soluzione analitica di Black-Scholes fornita
from numpy.lib.scimath import log, sqrt
from numpy import exp
from scipy.special import erf

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
    
######################################################
#CARICAMENTO DATI E GRAFICO
    
fig=plt.figure(figsize=(16,14))

plt.subplot(221)
x,y,err=np.loadtxt("ex_3.01/C_direct.out", usecols=(0, 1, 2), unpack=True)
plt.errorbar(x/1000,y,yerr=err)
plt.xlabel('N')
plt.ylabel('Prezzo')
plt.title("CALL DIRECTLY", fontweight="bold")
plt.hlines(call, 0, 100)
plt.grid(True)
print(" == DIRECTLY ==")
print("call: ", y[99], "+-",  err[99])

plt.subplot(222)
x,y,err=np.loadtxt("ex_3.01/P_direct.out", usecols=(0, 1, 2), unpack=True)
plt.errorbar(x/1000,y,yerr=err)
plt.xlabel('N')
plt.ylabel('Prezzo')
plt.title("PUT DIRECTLY ", fontweight="bold")
plt.hlines(put, 0, 100)
plt.grid(True)
print("put: ", y[99], "+-",  err[99])



fig=plt.figure(figsize=(16,14))

plt.subplot(221)
x,y,err=np.loadtxt("ex_3.01/C_discretized.out", usecols=(0, 1, 2), unpack=True)
plt.errorbar(x/1000,y,yerr=err)
plt.xlabel('N')
plt.ylabel('Prezzo')
plt.title("CALL DISCETIZED", fontweight="bold")
plt.hlines(call, 0, 100)
plt.grid(True)
print(" == DISCRETIZED ==")
print("call: ", y[99], "+-",  err[99])

plt.subplot(222)
x,y,err=np.loadtxt("ex_3.01/P_discretized.out", usecols=(0, 1, 2), unpack=True)
plt.errorbar(x/1000,y,yerr=err)
plt.xlabel('N')
plt.ylabel('Prezzo')
plt.title("PUT DISCRETIZED", fontweight="bold")
plt.hlines(put, 0, 100)
plt.grid(True)
print("put: ", y[99], "+-",  err[99])
```
