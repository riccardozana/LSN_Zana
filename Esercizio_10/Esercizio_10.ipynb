# Simulated Annealing 

Annealing, in italiano ricottura, è un processo utilizzato nella lavorazione dei metalli che consiste nel riscaldare il pezzo forgiato fino ad una temperatura molto elevata ed in seguito operare una lento raffreddamento in forno. Questa operazione di lento raffreddamento permette di ridurre la presenza di difetti nella stuttura del metallo, eliminando la presenza di stati metastabili formati e imprigionati nella struttura cristallina a causa di un raffreddamento più veloce. Si lascia di fatti la possibilità agli atomi di ridistribuirsi secondo la struttura che minimizza l'energia. 

Con Simulated Annealing si identifica quella classe di algoritmi metaeuristici che sfrutta un processo analogo per risolvere problemi di ottimizzazione. 
Nel nostro caso si è risolto il problema TSP con il SA ed è stato confrontato poi con il risultato ottenuto con GA. 
   
Una volta determinata la grandezza che si vuole minimizzare, nel nostro caso la lunghezza del percorso, questa viene utilizzata al posto dell'energia nel campionamento della distribuzione di probabilità di Boltzmann $P=\frac{e^{-\beta L}}{Z}$. 
Attraverso l'algoritmo Metropolis sono stati campionati circa $1.2 \times 10^6$ valori, con $\alpha=min(1, e^{-\beta(L_x-L_y)})$. Nel corso della simulazione la temperatura è è variata da un valore iniziale di $T_i=10$, fino a $T_f=0.0001$, secondo $T_y=T_x R$, con $R=0.99999$. 

Tale scelta è stata fatta in base a modifiche progressive al fine di avere una temperatura iniziale che permetta di sondare lo spettro dei cammini possibili e una discesa sufficientemente lenta per assicurare ergodicità durante la simulazione, specialmente a basse temperature. 


In seguito sono confrontati i risultati ottenuti nel caso di 32 città disposte in modo randomico e 32 città disposte all'interno del quadrato.

# Migrazione 

Nell'algoritmo genetico proposto nell'esericizio 9 e confrontato precedentemente con SA, è stata implementa la possibilità di effettuare delle migrazioni fra continenti. Per fare ciò sono state utilizzate le librerie MPI per effettuare calcolo parallelo sviluppando un programma secondo la tipologia SPMD (Single Program Multiple Data). Perciò la ricerca dell'individuo migliore è avvenuta secondo 4 processi indipendenti con una popolazione iniziale casuale e diversa per ognuno. Ognuno di questi processi, denominati continenti, evolve indipendentemente secondo le modalità descritte nell'esercitazione 9. Ogni $N$ generazioni, nel caso della simulazione corrente $N=200$, è stata effettuata una migrazione del miglior individuo di un continente in un altro continente estratto casualemente. L'obiettivo di questa mossa è quella di aumentare la diversità genetica degli individui in ogni continente e ridurre il tempo di computazione. Nel nostro caso, il numero di passi necessari ad ottenere il minimo globale fra i cammini del TSP è stato di $M=3874$ contro i $49923$ necessari nella simulazione senza migrazione. Contando il costo computazionale quadruplicato per ottenere $3874$ passi, la migrazione consente comunque di ridurre il costo computazionale rispetto all'algoritmo genetico base. 

In seguito sono mostrati i grafici relativi all'andamento medio della prima metà della popolazione e dell'individuo migliore per ogni processo. In alcuni casi è possibile osservare una variazione dell'individuo migliore in seguito alla migrazione. Infine è stato confrontanto il miglior cammino ottenuto con uno dei cammini già confrontati in precedenza.

```python
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(12,10))
fig.suptitle('MIGRAZIONE', fontsize=20, fontweight="bold")

plt.subplot(221)
plt.axvline(x=200, ymin=0, ymax=10, label="Migration", color="yellow")
for x in range(400, 4000, 200):
    plt.axvline(x, ymin=0, ymax=10, color="yellow")
x,L=np.loadtxt("ex_10.02/Square_LBest.0.out", usecols=(0, 1), unpack=True)
plt.plot(x,L, color="cyan", label="Best")
plt.title("Process 0", fontweight="bold")
plt.xlabel("N# Generations")
plt.ylabel("L")
plt.grid(True)
x,L=np.loadtxt("ex_10.02/Square_LAve.0.out", usecols=(0, 1), unpack=True)
plt.plot(x,L, color="purple", label="Average")
plt.legend()

plt.subplot(222)
plt.axvline(x=200, ymin=0, ymax=10, label="Migration", color="yellow")
for x in range(400, 4000, 200):
    plt.axvline(x, ymin=0, ymax=10, color="yellow")
x,L=np.loadtxt("ex_10.02/Square_LBest.1.out", usecols=(0, 1), unpack=True)
plt.plot(x,L, color="cyan", label="Best")
plt.title("Process 1", fontweight="bold")
plt.xlabel("N# Generations")
plt.ylabel("L")
plt.grid(True)
x,L=np.loadtxt("ex_10.02/Square_LAve.1.out", usecols=(0, 1), unpack=True)
plt.plot(x,L, color="purple", label="Average")
plt.legend()

plt.subplot(223)
plt.axvline(x=200, ymin=0, ymax=10, label="Migration", color="yellow")
for x in range(400, 4000, 200):
    plt.axvline(x, ymin=0, ymax=10, color="yellow")
x,L=np.loadtxt("ex_10.02/Square_LBest.2.out", usecols=(0, 1), unpack=True)
plt.plot(x,L, color="cyan", label="Best")
plt.title("Process 2", fontweight="bold")
plt.xlabel("N# Generations")
plt.ylabel("L")
plt.grid(True)
x,L=np.loadtxt("ex_10.02/Square_LAve.2.out", usecols=(0, 1), unpack=True)
plt.plot(x,L, color="purple", label="Average")
plt.legend()


plt.subplot(224)
plt.axvline(x=200, ymin=0, ymax=10, label="Migration", color="yellow")
for x in range(400, 4000, 200):
    plt.axvline(x, ymin=0, ymax=10, color="yellow")
x,L=np.loadtxt("ex_10.02/Square_LBest.3.out", usecols=(0, 1), unpack=True)
plt.plot(x,L, color="cyan", label="Best")
plt.title("Process 3", fontweight="bold")
plt.xlabel("N# Generations")
plt.ylabel("L")
plt.grid(True)
x,L=np.loadtxt("ex_10.02/Square_LAve.3.out", usecols=(0, 1), unpack=True)
plt.plot(x,L, color="purple", label="Average")
plt.legend()

fig=plt.figure(figsize=(12,6))

plt.subplot(121)
fig.suptitle('CONFRONTO CAMMINI', fontsize=20, fontweight="bold")
n,x,y =np.loadtxt("ex_10.02/Square_GlobalBestPath.out", usecols=(0, 1, 2), max_rows = 33, unpack=True)
plt.plot(x, y, '-o', color="green", linewidth=1)
plt.text(-1, y[0], "START-END", color="green", fontsize=12)
plt.title('Miglior Percorso',  fontweight="bold")
plt.xlabel("X")
plt.ylabel("Y")
plt.subplot(122)
n,x,y =np.loadtxt("ex_10.01/Square_BestPath_SA.out", usecols=(0, 1, 2), max_rows = 33, unpack=True)
plt.plot(x, y, '-o', color="Blue", linewidth=1)
plt.text(-1, y[0], "START-END", color="Red", fontsize=12)
plt.title('Miglior Percorso SA',  fontweight="bold")
plt.xlabel("X")
plt.ylabel("Y")
```
