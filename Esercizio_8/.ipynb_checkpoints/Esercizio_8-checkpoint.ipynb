### 8.01

Il primo punto dell'esercizio consiste nell'implementare il metodo VMC. Questo si basa sul principio variazionale, ovvero dato uno qualunque stato $|T\rangle$, l'energia dello stato $\langle {\hat H} \rangle_T $ sarà maggiore o uguale all'energia di ground state $\langle {\hat H} \rangle_{GS} $. 

Partendo da una famiglia di funzioni a uno o più parametri $\vec T$, [$\psi(\vec T)$], la ricerca della funzione d'onda di GS risulta essere un problema di ottimizzazione, cercando i paramentri che permettono di minimizzare il valore di $\langle {\hat H} \rangle_T $. 

Il valore di  $\langle {\hat H} \rangle_T $ può essere calcolato attraverso 
$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

dove l'ultimo termine dell'integrale, $\frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}$ prende il nome di Energia Locale. 

La famiglia di funzioni utilizzata nel caso in esame $$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}} $$
                               è proporzionale alla somma di due gaussiane centrate in $(x\pm\mu)$ e di larghezza $\sigma$, perciò dipendente da due parametri.
                               
Per il calcolo di $\langle {\hat H} \rangle_T$ si è anzitutto calcolata la normalizzazione $\frac{1}{\sqrt{N}}$ dello stato in funzione di $\mu$ e $\sigma$, come $N=\int dx |\Psi_T(x)|^2=2\sqrt{\pi}\sigma(1+e^{-\frac{\mu^2}{\sigma^2}})$. 

In seguito si è calcolato il valore di $\hat H|\Psi_T(x)\rangle$ come:

$$
    \hat H|\Psi_T(x)\rangle=(-\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2}+V(x))\Psi_T(x)= $$
$$-\frac{\hbar^2}{2mN}[(\frac{1}{\sigma^2}+\frac{x^2+\mu^2}{\sigma^4})(e^{-\frac{(x-\mu)^2}{2\sigma^2}}+e^{-\frac{(x+\mu)^2}{2\sigma^2}})-\frac{2x\mu}{\sigma^4}(e^{-\frac{(x-\mu)^2}{2\sigma^2}}-e^{-\frac{(x+\mu)^2}{2\sigma^2}})]+V(x)\Psi_T(x)= $$
$$
\Psi_T(x)[\frac{-\hbar^2}{2m\sigma^4}(x^2+\mu^2-\sigma^2-2x\mu\tanh\frac{(x-\mu)^2}{\sigma^2})+V(x)]
$$

con $\hbar=1$ e $m=1$.

Una volta determinate le espressioni analitiche delle quantità di interesse e stimati i valori di $\sigma$ e $\mu$, si è campionata la distribuzione di probabilità $|\Psi_T(x)|^2$ attraverso l'algoritmo di Metropolis per un numero di passi pari a $M=10^6$ e si è stimato il valore di $\langle {\hat H} \rangle_T$ e la corrispondente incertezza con il metodo a blocchi per un numero di blocchi $N=50$, ognuno di $L=2000$ passi.

### 8.02

Per l'ottimizzazione dei parametri è stata utilizata la tecnica del Simulated Annealing. 

La temperatura di partenza è stata scelta pari a $T_i=1$ e la temperatura finale pari a $T_f=0.0001$, con un rate di decrescita pari a $m=0.999$.
    Per ogni passo del SA, in totale 9207, sono stati estratti in un intorno circolare di raggio $r=0.1$ centrato in $(\sigma_{old} \mu_{old})$, due nuovi valori, $\sigma_{new}$ e $\mu_{new}$, ed è stato stimato il valore di $\langle {\hat H} \rangle_T=E_{new}$, ottenuto campionando $\Psi_T^{\sigma_{new},\mu_{new}}(x)$ nelle modalità spiegate in precedenza. Una volta determinato il valore di $E_{new}$ è stata calcolata la probabilità di accettazione come $p(x)=min(1, e^{-\beta(E_{new}-E_{old})})$ e sfruttando nuovamente l'algoritmo di Metropolis, si è accettata oppure scartata la nuova configurazione di $\sigma$ e $\mu$ fino al raggiungimento della temperatura finale.
    

Come si può notare dai grafici mostrati in seguito, i valori di $\sigma$ e $\mu$ sondano lo spazio delle possibili configurazioni ad alte temperature, stabilizzandosi sui valori ottimali al decrescere di quest'ultime.

Infine, si è confrontata la distribuzione di probabilità della funzione d'onda ottenuta con VMC con quella ottenuta analiticamente dalla soluzione matriciale dell'equazione di Schrodinger, mostrata nella sezione sottostante. 

Come si può notare dall'incompatibilità dell'energia di GS ottenuta con VMC con il valore analitico (primo grafico presentato), l'algoritmo di VMC presenta una criticità. La scelta della famiglia di funzioni da ottimizzate permette di raggiungere una buona approssimazione del GS solo se quest'ultimo appartiene alla famiglia stessa o presenta una forma molto simile. Perciò, come nel nostro caso, dove non si conosce la funzione di GS che si vuole approssimare, si ottiene una funzione ottimizzata simile alla funzione analitica ma non identica, e un valore dell'energia appena superiore, come atteso dal principio variazionale. 

```python
import numpy as np
import matplotlib.pyplot as plt

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2
hbar = 1
m = 1
a = 10
N = 1000 # number of iterations
# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)
# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2
CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal
# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)
# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)
# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)
print("Ground state energy: ", E[0])

# Plot a few things
plt.figure(figsize=(8,5))
scale = 0.3
pos=np.loadtxt("ex_8.02/prob.out", usecols=(0), unpack=True)
plt.hist(pos,100, (-3,3), density=True, color="yellow")
plt.plot(x,(psi[0])**2)
plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))
plt.show()
```

### 8.03
### Path integral projector method
Con Path integral projector method (PIGS) si intende una tecnica MonteCarlo che permette di stimare la funzione d'onda di GS. 
La tecnica sfrutta la propagazione per tempi immaginari $\tau$ dello stato quantisico iniziale $|\Psi(0)\rangle$; attraverso l'operatore di evoluzione temporale si ottiene che  $|\Psi(\tau)\rangle=e^{-\tau \hat H}|\Psi(0)\rangle=\sum_n c_n e^{-\tau E_n}|\Psi_n\rangle$, dove $|\Psi_n\rangle$ sono gli autostati dell'hamiltoniana. 

Come si può notare dall'ultimo membro, più lo stato iniziale non propagato è simile al GS, più il valore della proiezione su quest'ultimo sarà maggiore ($c_0$) e per ottenere lo stato GS con buona precisione non saranno necessari tempi immaginari $\tau$ molto lunghi. D'altra parte, indipendentemente dallo stato iniziale (salvo se ortogonale al GS) nel limite di $\tau \to \infty$, "sopravvive" solamente il termine propozionale al GS. 

Si è utilizzata la base delle posizioni $|\vec r \rangle$, dove 
$|\Psi(\vec r, \tau)\rangle=\int d\vec r' \langle \vec r| e^{-\tau \hat H} |\vec r' \rangle |\Psi(\vec r', 0)\rangle  $ a partire dalla quale si è decomposto il propagatore secondo la decomposizione di Trotter in $M$ fattori $e^{-\frac{\tau}{M}\hat H}$. Per il calcolo del risultato del propagatore si sono fatte due approssimazioni: 
- $\tau$ è finito
- si è utilizzata l'approssimazione primitiva per effettuare il calcolo del propagatore fattorizzato in $e^{\delta \tau(\hat T + \hat V)}$, dove $\delta\tau=\frac{\tau}{M}<<1$

La funzione d'onda di GS risulta essere il risultato delle $M$ convoluzioni a tempi immaginari e il valore medio dell'osservabile sul GS è equivalente alla media secondo la distribuzione di probabilità canonica di $2M+1$ particelle di una catena lineare polimerica, perciò risulta sufficiente campionare la distribuizione di probabilità congiunta delle $2M+1$ particelle. 

Il metodo in esame può essere considerato "esatto" perchè modificando il tempo di propagazione immaginario $\tau$ e il numero delle fattorizzazioni $M$, è possibile ridurre a piacere l'errore dovuto alle approssimazioni effettuate, rendendolo minore dell'incertezza statistica dovuta al campionamento delle distribuzioni di probabilità con MC. 

In seguito sono mostrati i risultati ottenuti con PIGS.

Una volta inserite le quantità di input, la simulazione effettua 3 algoritmi Metropolis per la traslazione rigida del polimero, il moto delle code e il brownian bridge.
I valori di input sono stati settati ad ogni simulazione per ottenere una percentuale dei tre valori di accettazione compresa fra il $45\%-55\%$.

Le simulazioni sono state fatte con due funzioni d'onda di input, la funzione costante $\Psi(x)=1$ e quella precedentemente ottimizzata $\Psi_T^{\sigma,\mu}(x)$. 
In seguito sono mostrati i risultati comparati con la soluzione analitica matriciale mostrata in precedenza, per valori di $\tau$ differenti.
Gli istogrammi sono stati riempiti con i valori compresi fra 120 e 180 timeslices. In questi intervalli posizionati centralmente la funzione d'onda risulta essere maggiormente precisa perchè il bra e il ket risultano maggiormente simili allo stato fondamentale.
Nella zona centrale infatti il calcolo del potenziale V viene effettuato su due stati proiettati per un tempo $\tau/2$ ($<T\tau/2|V|\tau/2>$) mentre nelle timeslices laterali su $<\tau|V|0>$ e viceversa, rendendo il calcolo del potenziale molto impreciso, perchè una dei due stati è la funzione di prova, molto diversa dallo stato fondamentale; nella zona centrale invece lo stato fondamentale e quello ottenuto tramite la proiezione risultano essere statisticamente indistinguibili

Come si può notare, per $\Psi(x)=1$ sono necessari tempi di propagazione maggiori per avere uno stato correttamente proiettato sul GS, proprio a causa della sua minore proiezione iniziale $C_0$, rispetto allo stato già ottimizzato. 

Per entrambe le simulazioni è stato scelto $M=300$ mentre $\tau=12.0$ nel caso di funzione costante e $\tau=1, 4, 8$ nel caso di quella ottimizzata. La compatibilità con il GS è stata stimata visivamente per valori di $\tau$ inferiori fino ad ottenere i valori graficati; per lo stato ottimizzato la compatibilità con il GS analitico risulta buona fino a $\tau=4$.

```python

N = 400 # number of iterations
# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)
CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

Em,psi = np.linalg.eigh(H)
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)


fig=plt.figure(figsize=(20,20))
fig.suptitle('PIGS', fontsize=20, fontweight="bold")

plt.subplot(221)
x,k,kerr=np.loadtxt("ex_8.03/kinetic.dat", usecols=(0, 1, 2), unpack=True)
v,verr=np.loadtxt("ex_8.03/potential.dat", usecols=(1, 2), unpack=True)
E=k+v
Eerr =np.sqrt(kerr*kerr+verr*verr)
x= range(E.size)
plt.plot(x, E, color="cyan", label = "Energy PIGS", linewidth=0.5)
plt.hlines(Em[0], x[0], x[-1], color = "green", label = 'Energy Matrix Method')
plt.hlines(Efinal[-1], x[0], x[-1], color = 'red', label='Energy VMC')
plt.errorbar(x, v, verr, label = "Potential PIGS")
plt.title("Energy PIGS - Constant Function", fontweight="bold")
plt.xlabel("M")
plt.ylabel("Energy")
plt.legend()
plt.grid(True)

plt.subplot(222)
x,prob,perr=np.loadtxt("ex_8.03/probability.dat", usecols=(0, 1, 2), unpack=True) 
plt.errorbar(x, prob-((psi[0])**2), perr, label = "Constant T=12")
plt.hlines(0, x[0], x[-1], label = "GS")
plt.title("GS Functions comparison - Constant", fontweight="bold")
plt.xlabel("x")
plt.ylabel("$|\psi|^2$")
plt.legend()

plt.grid(True)

plt.subplot(223)
x,k,kerr=np.loadtxt("ex_8.03/kinetic_var_8.dat", usecols=(0, 1, 2), unpack=True)
v,verr=np.loadtxt("ex_8.03/potential_var_8.dat", usecols=(1, 2), unpack=True)
E=k+v
Eerr =np.sqrt(kerr*kerr+verr*verr)
x= range(E.size)
plt.plot(x, E, color="cyan", label = "Energy PIGS", linewidth=0.5)
plt.hlines(Em[0], x[0], x[-1], color = "green", label = 'Energy Matrix Method')
plt.hlines(Efinal[-1], x[0], x[-1], color = 'red', label='Energy VMC')
plt.errorbar(x, v, verr, label = "Potential PIGS")
plt.title("Energy PIGS - VMC function", fontweight="bold")
plt.xlabel("M")
plt.ylabel("Energy")
plt.legend()
plt.grid(True)

plt.subplot(224)
z = np.linspace(-a/2, a/2, N)
x,prob8,perr8=np.loadtxt("ex_8.03/probability_var_8.dat", usecols=(0, 1, 2), unpack=True)
plt.errorbar(x, prob8-((psi[0])**2), perr8, label="VMC T=8")
plt.hlines(0, x[0], x[-1], label = "GS")
plt.title("GS Functions comparison - VMC", fontweight="bold")
plt.xlabel("x")
plt.ylabel("$|\psi|^2$")
plt.grid(True)
plt.legend()

plt.show()



fig=plt.figure(figsize=(12,15))

plt.subplot(311)
x,prob4,perr4=np.loadtxt("ex_8.03/probability_var_4.dat", usecols=(0, 1, 2), unpack=True)
x,prob1,perr1=np.loadtxt("ex_8.03/probability_var_1.dat", usecols=(0, 1, 2), unpack=True)
plt.errorbar(x, prob8, perr8, label = "T=8")
plt.errorbar(x, prob4, perr4, label = "T=4")
plt.errorbar(x, prob1, perr1, label = "T=1")
plt.legend()
plt.title("Tau comparison - VMC", fontweight="bold")
plt.xlabel("x")
plt.ylabel("$|\psi|^2$")
plt.grid(True)


plt.subplot(312)
plt.errorbar(x, prob4-((psi[0])**2), perr4, label = "T=4", color="orange")
plt.hlines(0, x[0], x[-1], label = "GS Matrix Method")
plt.title("Tau 4 - VMC", fontweight="bold")
plt.xlabel("x")
plt.ylabel("$|\psi|^2$")
plt.grid(True)
plt.legend()

plt.subplot(313)
plt.errorbar(x, prob1-((psi[0])**2), perr1, label = "T=1", color="green")
plt.hlines(0, x[0], x[-1], label = "GS Matrix Method")
plt.title("Tau 1 - VMC", fontweight="bold")
plt.xlabel("x")
plt.ylabel("$|\psi|^2$")
plt.grid(True)
plt.legend()

plt.show()
```
